```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])
y = np.array([0,0,0,1])

np.random.seed(1)
w = np.random.rand(2)
b = np.random.rand(1)
lr=0.1
epochs = 10

def step_function(z):
    return np.where(z>=0,1,0)

for epoch in range(epochs):
    print(f"Epoch : {epoch+1}")
    for i in range(len(x)):
        z=np.dot(x[i], w)+b
        y_pred = step_function(z)

        error = y[i] - y_pred

        w += lr*x[i]*error
        b += lr*error
        print(f"Weight: {w} , Predicted : {y_pred}")

x1 = np.linspace(-0.5,1.5,100)
x2 = -(w[0]*x1+b)/w[1]

plt.scatter(x[:,0] , x[:,1] ,c=y , cmap = 'bwr' , s= 100)
plt.plot(x1,x2)
plt.show()
```
